```python
import pandas
df = pandas.read_csv('names.csv',
                    index_col='ID')

print(df)
```

```python
df1 = pandas.DataFrame(df)
df1
```

```python
d1=df1.iloc[:,2]
d1
```

```python
d2=df1.iloc[:,1]
d2
```

```python
import matplotlib.pyplot as plt
# An "interface" to matplotlib.axes.Axes.hist() method
df1, bins, patches = plt.hist(x=d1, bins=20, color='#0504aa',
                            alpha=0.7, rwidth=20)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Weight in kilograms')
plt.ylabel('Frequency')
plt.title('Weights from my data for this lecture')
#plt.text(10, 10, 'Sila you did well - a very nice histogram')
maxfreq = df1.max()
plt.savefig('Sila_histogram20.png', dpi=300)
```

```python
import matplotlib.pyplot as plt
# An "interface" to matplotlib.axes.Axes.hist() method
df1, bins, patches = plt.hist(x=d2, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=2)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Height in meters')
plt.ylabel('Frequency')
plt.title('Height from my data for this lecture')
#plt.text(10, 10, 'Sila you did well - a very nice histogram')
maxfreq = df1.max()
plt.savefig('Sila_histogram_height.png', dpi=300)
```
